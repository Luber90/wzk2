```python
%matplotlib inline
import numpy as np
import random
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
img = mpimg.imread('szyfr.png')
for i in range(len(img)):
    for j in range(len(img[i])):
        if(img[i, j] != 1): img[i, j] = 0
plt.imshow(img, cmap='gray')
```

```python
img1 = np.empty((100, 200))
img2 = np.empty((100, 200))
for i in range(len(img)):
    for j in range(len(img[i])):
        which = random.randint(0, 1)
        if(img[i, j]==0):
            if(which==0):
                img1[i, j*2]=1
                img1[i, j*2+1]=0
                img2[i, j*2]=0
                img2[i, j*2+1]=1
            else:
                img1[i, j*2]=0
                img1[i, j*2+1]=1
                img2[i, j*2]=1
                img2[i, j*2+1]=0
        else:
            if(which==1):
                img1[i, j*2]=1
                img1[i, j*2+1]=0
                img2[i, j*2]=1
                img2[i, j*2+1]=0
            else:
                img1[i, j*2]=0
                img1[i, j*2+1]=1
                img2[i, j*2]=0
                img2[i, j*2+1]=1
```

```python
plt.imshow(img1, cmap='gray')
```

```python
plt.imshow(img2, cmap='gray')
```

```python
result = np.ones((100, 200))
for i in range(len(img1)):
    for j in range(len(img1[i])):
        if(img1[i, j]==0 or img2[i, j]==0):
            result[i, j] = 0
plt.imshow(result, cmap='gray')
```
